Website Visits

```python
import numpy as np
```

```python
pages = np.array([
    "Home",
    "About",
    "Home",
    "Contact",
    "Shop",
    "Home",
])

unique, counts = np.unique(pages, return_counts=True)

most = unique[counts==np.max(counts)]

least = unique[counts==np.min(counts)]

print("Most visited pages: ", most)
print("Least visited pages: ",least)
```
